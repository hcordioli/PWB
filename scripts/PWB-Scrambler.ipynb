# Load dataset

```python
pwbdatafile = '..\\data\\pwb-data.csv'
pwbds = pd.read_csv(pwbdatafile)
pwbds.drop(['time'],axis=1,inplace=True)
pwbds.shape
```

```python
len(pwbds.loc[pwbds.y==1])
```

# Isola cada s√©rie

```python
brks = pwbds.loc[pwbds.y==1]
```

```python
start=0
end=260
tmp = pwbds.iloc[start:end+1,:]
```

```python
brkseries = list()
start=0
for brk_idx, brk in brks.iterrows():
    tmp = pwbds.iloc[start:brk_idx+1,:]
    brkseries.append(tmp)
    start = brk_idx+1
```

# Augment original dataset randomly

```python
for i in range (2):
    new_series = random.sample(range(0, len(brks)), len(brks))
    for new_serie in new_series:
        pwbds = pwbds.append(brkseries[new_serie])
```
